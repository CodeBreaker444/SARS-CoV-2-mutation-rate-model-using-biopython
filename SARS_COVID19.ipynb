```python
def cb_calculate_similarity(genome1, genome2):
    #print(genome1,genome2)
    pa=PairwiseAligner()
    pairwise_score=pa.score(genome1, genome2)
    return pairwise_score,len(genome1)+len(genome2)
    
```

```python
perm_regions_names=list(itertools.permutations(['covid19_italy', 'covid19_kerala', 'covid19_nepal', 'covid19_wuhan', 'covid19_usa'],2))
perm_regions_names=numpy.array([*perm_regions_names]).tolist()
perm_regions=list(itertools.permutations([covid19_italy, covid19_kerala, covid19_nepal, covid19_wuhan, covid19_usa],2))
```

```python
rate_mutation=[]
for i in range(0,len(perm_regions)):
    a=perm_regions[i][0]
    b=perm_regions[i][1]
    a,b=cb_calculate_similarity(a,b)
    rate_mutation.append((a/(b/2))*100)
    print(rate_mutation)
```

```python
for i in range(0,len(rate_mutation)):
    perm_regions_names[i].append(rate_mutation[i])
    perm_regions_names[i].append(100-rate_mutation[i])
```
